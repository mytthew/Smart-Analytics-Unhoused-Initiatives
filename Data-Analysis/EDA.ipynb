```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Bring in our data
file_path = '..//Resources/Clean Data/cleaned_coc_df.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Preview the dataframe
df.head()
```

# AGE 

```python
# Create an Age Dataframe for OVERALL HOMELESS

age_columns = [
    'State',
    'CoC Name',
    'Overall Homeless',
    'Overall Homeless - Under 18',
    'Overall Homeless - Age 18 to 24',
    'Overall Homeless - Age 25 to 34',
    'Overall Homeless - Age 35 to 44',
    'Overall Homeless - Age 45 to 54',
    'Overall Homeless - Age 55 to 64',
    'Overall Homeless - Over 64'
]

# Extract the specific columns
age_df = df[age_columns]

age_df.head()
```

```python
# Calculate the sum of each column individually
sum_overall_homeless = age_df['Overall Homeless'].sum()
sum_under_18 = age_df['Overall Homeless - Under 18'].sum()
sum_age_18_to_24 = age_df['Overall Homeless - Age 18 to 24'].sum()
sum_age_25_to_34 = age_df['Overall Homeless - Age 25 to 34'].sum()
sum_age_35_to_44 = age_df['Overall Homeless - Age 35 to 44'].sum()
sum_age_45_to_54 = age_df['Overall Homeless - Age 45 to 54'].sum()
sum_age_55_to_64 = age_df['Overall Homeless - Age 55 to 64'].sum()
sum_age_over_64 = age_df['Overall Homeless - Over 64'].sum()


# Format the sums nicely
formatted_sums = {
    'Overall Homeless': '{:,.0f}'.format(sum_overall_homeless),
    'Overall Homeless - Under 18': '{:,.0f}'.format(sum_under_18),
    'Overall Homeless - Age 18 to 24': '{:,.0f}'.format(sum_age_18_to_24),
    'Overall Homeless - Age 25 to 34': '{:,.0f}'.format(sum_age_25_to_34),
    'Overall Homeless - Age 35 to 44': '{:,.0f}'.format(sum_age_35_to_44),
    'Overall Homeless - Age 45 to 54': '{:,.0f}'.format(sum_age_18_to_24),
    'Overall Homeless - Age 55 to 64': '{:,.0f}'.format(sum_age_25_to_34),
    'Overall Homeless - Over 64': '{:,.0f}'.format(sum_age_35_to_44)
}

# Create a DataFrame to store the formatted sums
age_sum_df = pd.DataFrame(formatted_sums.items(), columns=['Column', 'Sum'])

# Display the DataFrame
age_sum_df
```

```python
# Format as Percentages 

# Calculate the sum of the "Overall Homeless" column
total_overall_homeless = age_df['Overall Homeless'].sum()

# Calculate the percentage for each sum and format as percentage
percentage_overall_homeless = {
    'Overall Homeless - Under 18': '{:.2%}'.format(sum_under_18 / total_overall_homeless),
    'Overall Homeless - Age 18 to 24': '{:.2%}'.format(sum_age_18_to_24 / total_overall_homeless),
    'Overall Homeless - Age 25 to 34': '{:.2%}'.format(sum_age_25_to_34 / total_overall_homeless),
    'Overall Homeless - Age 35 to 44': '{:.2%}'.format(sum_age_35_to_44 / total_overall_homeless),
    'Overall Homeless - Age 45 to 54': '{:.2%}'.format(sum_age_45_to_54 / total_overall_homeless),
    'Overall Homeless - Age 55 to 64': '{:.2%}'.format(sum_age_55_to_64 / total_overall_homeless),
    'Overall Homeless - Over 64': '{:.2%}'.format(sum_age_over_64 / total_overall_homeless)
}

# Create a DataFrame to store the percentages
age_percentage_df = pd.DataFrame(percentage_overall_homeless.items(), columns=['Column', 'Percentage of Overall Homeless'])

# Display the DataFrame
age_percentage_df
```

```python
# Create Pie Chart

# Extract labels and percentages from the DataFrame
labels = age_percentage_df['Column']
percentages = age_percentage_df['Percentage of Overall Homeless']

# Convert percentages from string to float for plotting
percentages_float = percentages.str.rstrip('%').astype(float)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages_float, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Grouped by Age')
plt.suptitle('Percent of Overall Homeless')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.tight_layout() 
plt.gcf().set_size_inches(10, 6) 
plt.savefig('Percent of Overall Homeless - Grouped by Age')
plt.show()
```

As we can see from this chart there is not one single age group that dominates the overall homeless population, however the age group consisiting of 35-44 year old individuals does take up the largest percent at 19.4%.

The next two highest aren't far behind. Under 18 represents 18.2% and Ages 25-34 represent 18.1%. 

Individuals age 18 to 24 account for just 7.7% of the overall homeless population.

# VETERANS

```python
# VETERANS Dataframe

vets_columns = [
    'State',
    'CoC Name',
    'Overall Homeless',
    'Overall Homeless Veterans',
    'Overall Homeless Veterans - Female',
    'Overall Homeless Veterans - Male'
]

vets_df = df[vets_columns]

vets_df.head()
```

```python
# Calculate the sum of the "Overall Homeless" column
total_overall_homeless = vets_df['Overall Homeless'].sum()
total_overall_homeless_vets = vets_df['Overall Homeless Veterans'].sum()

# Calculate sums for each gender category
total_overall_female_homeless_vets = vets_df['Overall Homeless Veterans - Female'].sum()
total_overall_male_homeless_vets = vets_df['Overall Homeless Veterans - Male'].sum()


# Calculate the percentage for each sum and format as percentage
percentage_overall_homeless = {
    'Homeless Vets - % of Total Homeless': '{:.2%}'.format(total_overall_homeless_vets / total_overall_homeless),
    'Homeless Female Vets - % of Total Homeless Vets': '{:.2%}'.format(total_overall_female_homeless_vets / total_overall_homeless_vets ),
    'Homeless Male Vets - % of Total Homeless Vets': '{:.2%}'.format(total_overall_male_homeless_vets / total_overall_homeless_vets ),
}

# Create a DataFrame to store the percentages
vets_percentage_df = pd.DataFrame(percentage_overall_homeless.items(), columns=['Column', 'Percentage of Overall Homeless'])

# Display the DataFrame
vets_percentage_df
```

```python
# Create Pie Chart

# Filter the DataFrame to include only the rows for "Homeless Female Vets" and "Homeless Male Vets"
vets_df2 = vets_percentage_df.loc[1:2]

# Extract labels and percentages from the filtered DataFrame
labels = vets_df2['Column']
percentages = vets_df2['Percentage of Overall Homeless']

# Convert percentages from string to float for plotting
percentages_float = percentages.str.rstrip('%').astype(float)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages_float, labels=labels, autopct='%1.2f%%', startangle=140)
plt.title('Homeless Vets by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.tight_layout()  
plt.gcf().set_size_inches(10, 6) 
plt.savefig('Homeless Vets by Gender')
plt.show()
```

As we can see here the overwhelming majority of homeless vets (89%) are Male.

# RACE 

```python
# RACE Dataframe

races_columns = [
    'State',
    'CoC Name',
    'Overall Homeless',
    'Overall Homeless Veterans',
    'Overall Homeless - Hispanic/Latin(o)(a)(x)',
    'Overall Homeless - White',
    'Overall Homeless - Black, African American, or African',
    'Overall Homeless - Asian or Asian American',
    'Overall Homeless - American Indian, Alaska Native, or Indigenous',
    'Overall Homeless - Native Hawaiian or Other Pacific Islander',
    'Overall Homeless - Multiple Races'
]

races_df = df[races_columns]

races_df.head()
```

```python
# Calculate the sum of the "Overall Homeless" column
total_overall_homeless = df['Overall Homeless'].sum()
total_overall_homeless_racesdf = races_df['Overall Homeless'].sum()

# Calculate sums for each gender category
total_overall_latinos = races_df['Overall Homeless - Hispanic/Latin(o)(a)(x)'].sum()
total_overall_white = races_df['Overall Homeless - White'].sum()
total_overall_african_american = races_df['Overall Homeless - Black, African American, or African'].sum()
total_overall_asian = races_df['Overall Homeless - Asian or Asian American'].sum()
total_overall_indigenous = races_df['Overall Homeless - American Indian, Alaska Native, or Indigenous'].sum()
total_overall_islander = races_df['Overall Homeless - Native Hawaiian or Other Pacific Islander'].sum()
total_overall_multiple_races = races_df['Overall Homeless - Multiple Races'].sum()

# Calculate the percentage for each sum and format as percentage
percentage_overall_homeless = {
    'Latinos': '{:.2%}'.format(total_overall_latinos / total_overall_homeless),
    'White': '{:.2%}'.format(total_overall_white / total_overall_homeless),
    'African American': '{:.2%}'.format(total_overall_african_american / total_overall_homeless),
    'Asian': '{:.2%}'.format(total_overall_asian / total_overall_homeless),
    'Indigenous': '{:.2%}'.format(total_overall_indigenous / total_overall_homeless),
    'Islander': '{:.2%}'.format(total_overall_islander / total_overall_homeless),
    'Multiple Races': '{:.2%}'.format(total_overall_multiple_races / total_overall_homeless),
}

# Create a DataFrame to store the percentages
races_percentage_df = pd.DataFrame(percentage_overall_homeless.items(), columns=['Column', 'Homeless By Race'])

# Display the DataFrame
races_percentage_df
```

```python
# Create Pie Chart

# Extract labels and percentages from the DataFrame
labels = races_percentage_df['Column']
percentages = races_percentage_df['Homeless By Race']

# Convert percentages from string to float for plotting
percentages_float = percentages.str.rstrip('%').astype(float)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages_float, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Grouped by Race')
plt.suptitle('Percent of Overall Homeless')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.tight_layout() 
plt.gcf().set_size_inches(10, 6) 
plt.savefig('Percent of Overall Homeless - Grouped by Race')
plt.show()
```

We can see here that White is the race with the most overall homeless, representing 39% of the total homeless population in our dataset. The next two highest are African American (29%) and Latinos (21.6%). Islanders (1.3%) represents the smallest percentage of overall homeless.

# STATES WITH MOST AND LEAST HOMELESS 

```python
# Grouping by 'State' and summing the 'Overall Homeless' counts
top_states = df.groupby('State')['Overall Homeless'].sum()

# Sorting the result in descending order and selecting the top 10
top_10_states = top_states.sort_values(ascending=False).head(10)

# Creating a bar chart
top_10_states.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Adding title and labels
plt.title('Top 10 States with the Most Overall Homeless Counts')
plt.xlabel('State')
plt.ylabel('Overall Homeless Count')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.tight_layout()
plt.savefig('States With the Most Homeless')
plt.show()
```

California and New York are by far the two states with the most amount of homeless persons. 

```python
# Grouping by 'State' and summing the 'Overall Homeless' counts
bottom_states = df.groupby('State')['Overall Homeless'].sum()

# Sorting the result in descending order and selecting the top 10
bottom_10_states = bottom_states.sort_values(ascending=True).head(10)

# Creating a bar chart
bottom_10_states.plot(kind='bar', figsize=(10, 6), color='skyblue')

# Adding title and labels
plt.title('Top 10 States with the Least Overall Homeless Counts')
plt.xlabel('State')
plt.ylabel('Overall Homeless Count')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.tight_layout()
plt.savefig('States With the Least Homeless')
plt.show()
```
